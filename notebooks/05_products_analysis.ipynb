This progression suggests a structural shift in demand composition as the platform matured.

Notably, the leading category in 2016 (Furniture) generated substantially less revenue than the top categories in 2017 and 2018. To further strengthen the analysis, we examine the monthly revenue trends of the five categories that generated the highest cumulative revenue over the entire period.

```python
df2 = pd.read_csv(path + 'monthly_revenue_by_product_category.csv')

# Create datetime column
df2['date'] = pd.to_datetime(
    dict(year=df2['year'], month=df2['month'], day=1))


# Sort chronologically
df2 = df2.sort_values(['category', 'date'])

plt.figure()

for category, group in df2.groupby('category'):
    plt.plot(group['date'], group['revenue'], label=category)

plt.xlabel('Month')
plt.ylabel('Revenue (R$)')
plt.title('Monthly Revenue by Category')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

The monthly revenue trajectories of the five highest revenue-generating categories reveal a clear upward trend over time, consistent with the overall platform expansion identified in the prior revenue_analysis notebook.

Across categories, revenue generally increases as time progresses, reflecting both marketplace growth and rising transaction activity.

A noticeable seasonal pattern emerges beginning in 2017. In particular, most categories exhibit a pronounced spike in revenue around November 2017, followed by a decline in subsequent months. This pattern is consistent with seasonal demand effects, potentially driven by holiday-related consumption and promotional cycles.

An exception to this pattern is Health & Beauty, which, while also displaying a November spike, continues to grow afterward rather than reverting to pre-spike levels. This sustained upward trajectory suggests that the category may have experienced structural demand growth beyond seasonal fluctuations.

Unlike most categories, Computer Accessories and Watches exhibit a second noticeable spike in order volume in February 2018 and May 20218, respectively.

This pattern may indicate that demand in these segments is influenced by additional seasonal or event-driven factors beyond year-end holiday consumption.

Possible explanations include:

- Post-holiday purchasing behavior, where consumers use gift money or exchange products.
- Back-to-school or academic cycle effects, particularly relevant for technology-related products such as computer accessories.
- Product upgrade cycles, which are more common in electronics and wearable devices.
- Promotional campaigns or sales events occurring early in the year.
- Specific calendar events, which affect mostly luxury-adjacent goods.

The fact that this secondary spike is concentrated in specific categories suggests that seasonality is not uniform across the marketplace. Instead, demand patterns appear to be category-dependent, reflecting differences in consumer purchasing motives and product lifecycle dynamics.

Overall, the evidence indicates that both platform expansion and seasonal effects contribute to revenue dynamics, with category-specific growth patterns suggesting heterogeneous demand evolution across segments.
